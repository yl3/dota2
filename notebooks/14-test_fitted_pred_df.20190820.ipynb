## Load and fit the TI9 group stages dataset.

```python
JSON_FILE = "data/raw/premium_matches.2019-08-19.json.gz"
with gzip.open(JSON_FILE, 'rb') as fh:
    matches = src.load.MatchDF.from_json(json.load(fh)['data'])
```

## Test `fitted_pred_df()` and `fitted_skills_mat()`.

```python
ms_in_a_year = 365 * 24 * 60 * 60 * 1000
gp_model = src.models.gp.SkillsGPMAP(
    initial_skills=None,
    initial_radi_adv=None,
    players_mat=matches.players_mat,
    start_times=matches.df.startTimestamp,
    radiant_win=matches.df.radiantVictory,
    cov_func_name='exponential',
    cov_func_kwargs={'scale': 1.25 * ms_in_a_year},
    radi_prior_sd=3.0,
    logistic_scale=3.0
)
```

```python
gp_model.fit()
```
