## Explore bias checking in predictions.

```python
def match_pred_df_bias(match_pred_df,
                       win_prob_breaks=np.arange(0.0, 1.1, 0.1)):
    """Compute prediction bias in a match prediction data frame.
    
    Match prediction data frame must have columns 'radiantVictory' and
    'pred_win_prob'
    """
    win_prob_interval = \
        pd.cut(match_pred_df.pred_win_prob, win_prob_breaks,
               include_lowest=True)
    grouped = match_pred_df.groupby(win_prob_interval)
    res_df = pd.DataFrame(
        [pd.Series(
            src.stats.win_oe_pval_series(df.pred_win_prob, df.radiantVictory),
            name=name)
         for name, df in grouped])
    return res_df
```

### Create helper functions.

```python
def _match_entry_to_hovertext(match, prediction, cur_skills):
    """Convert one entry of a match with predictions to hovertext."""
    radi_skills = cur_skills.loc[match.radiant_players]
    dire_skills = cur_skills.loc[match.dire_players]
    radi_roster = "</br>".join(
        ["{} ({}) | <b>{:.2f}</b>".format(name, id, skill)
         for name, id, skill in zip(match.radiant_nicknames,
                                    match.radiant_players, radi_skills)])
    dire_roster = "</br>".join(
        ["{} ({}) | <b>{:.2f}</b>".format(name, id, skill)
         for name, id, skill in zip(match.dire_nicknames, match.dire_players,
                                    dire_skills)])
    title = "<b>{} | {}</b>".format(
        match.startDate.strftime("%Y-%m-%d %H:%M"), match.name)
    subtitle = "Predicted win prob: {:.2f}".format(prediction.pred_win_prob)
    radi_title = "<b>{} ({}) | {:.3f}</b>".format(
        match.radiant_name, match.radiant_valveId, prediction.radi_skill)
    dire_title = "<b>{} ({}) | {:.3f}</b>".format(
        match.dire_name, match.dire_valveId, prediction.dire_skill)
    output = "</br>".join([
        title, "",
        subtitle, "",
        radi_title,
        radi_roster, "",
        dire_title,
        dire_roster, "",
        "radi_adv: <b>{:.2f}</b>".format(prediction.radi_adv)
    ])
    return(output)

def matches_to_hover_info(matches, predictions_df, skills_mat):
    """Create hovertext for plotly data points."""
    hovertext = [_match_entry_to_hovertext(m[1], p[1], s[1])
                 for m, p, s in zip(matches.iterrows(),
                                    predictions_df.iterrows(),
                                    skills_mat.iterrows())]
    return pd.Series(hovertext, index=predictions_df.index)

def side_outcome_to_symbol(is_radi, radi_win):
    """Compute Plotly symbol based on side and outcome."""
    symbol = np.where(is_radi,
                      np.where(radi_win, "triangle-up", "triangle-down"),
                      np.where(radi_win, "triangle-down-open",
                               "triangle-up-open"))
    return symbol

def data_to_graph_obj(time, y, two_sds, is_radi, radi_win, hovertext, 
                      name=None):
    """Convert data into a Plotly graph object."""
    df = pd.DataFrame({'x': time, 'y': y, 'err': two_sds, 'is_radi': is_radi,
                       'radi_win': radi_win})
    df = df.sort_values('x')
    marker = {'symbol': side_outcome_to_symbol(df.is_radi, df.radi_win),
              'size': 8, 'opacity': 0.8}
    line = {'width': 1}
    data = go.Scatter(x=df.x, y=df.y, marker=marker, mode='lines+markers',
                      line=line, name=name, showlegend=True,
                      hovertext=hovertext, hoverinfo="text")
    return data
```
